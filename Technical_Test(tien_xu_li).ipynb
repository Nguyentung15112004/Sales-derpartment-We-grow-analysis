ở phần này do dữ liệu thiếu là rất nhiều nên em khuyến nghịc dùng học máy( ở đây em dùng desecsion tree classfication) các bước chi tiết e có note ở code.

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

columns_missing = ['Marketer 2']  # Các cột cần dự đoán
feature_cols = ["Channel", "Chiến dịch", "Content", "Khách hàng"]  # Các cột đặc trưng

# Mã hóa tất cả các cột đặc trưng thành số
encoders = {}
for col in feature_cols:
    le = LabelEncoder()
    df2[col] = le.fit_transform(df2[col].astype(str))  # Chuyển đổi cột thành số
    encoders[col] = le  # Lưu encoder để sử dụng sau

# Xử lý từng cột bị thiếu dữ liệu
for col_missing in columns_missing:
    print(f"Đang xử lý cột: {col_missing}")

    # Chia dữ liệu thành hai phần
    df_cleaned_train = df2[df2[col_missing].notnull()]  # Dữ liệu đầy đủ
    df_cleaned_missing = df2[df2[col_missing].isnull()]  # Dữ liệu thiếu

    # Dữ liệu đầu vào (đã được mã hóa)
    X_train = df_cleaned_train[feature_cols]
    y_train = df_cleaned_train[col_missing]

    # Mã hóa cột cần dự đoán
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train.astype(str))

    # Huấn luyện mô hình Decision Tree
    model = DecisionTreeClassifier(max_depth=10, random_state=42)
    model.fit(X_train, y_train_encoded)

    # Dự đoán giá trị bị thiếu
    X_missing = df_cleaned_missing[feature_cols]
    predicted_encoded_values = model.predict(X_missing)

    # Giải mã từ số về dạng chữ
    df_cleaned_missing[col_missing] = label_encoder.inverse_transform(predicted_encoded_values)

    # Cập nhật lại df2 với giá trị đã điền
    df2.loc[df2[col_missing].isnull(), col_missing] = df_cleaned_missing[col_missing]

# Kiểm tra kết quả
print("Số giá trị NaN sau khi xử lý:")
print(df2[columns_missing].isnull().sum())  # Kiểm tra còn NaN không
```

Mã hóa lại cột về dạng chữ


```python
for col in feature_cols:
    if col in encoders:  # Kiểm tra xem cột có trong danh sách mã hóa không
        df2[col] = encoders[col].inverse_transform(df2[col])
```
